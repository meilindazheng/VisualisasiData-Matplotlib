## Bar Plot / Bar Chart 
Sesi ini akan berfokus ke cara membuat barplot dengan matplotlib

### Import Module

```python
#magic command untuk matplotib
%matplotlib inline
```

```python
#mengimport module yang digunakan 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
#mencetak versi matplotlib dan numpy yang sedang digunakan
print(matplotlib.__version__)
print(np.__version__)
```

### Simple Bar Plot
- plt.bar() akan membutuhkan 2 parameter, parameter pertama akan digambarkan melalui koordinat x dan parameter kedua melalui koordinat y
- plt.xlabel() akan memberikan label pada koordinat x
- plt.ylabel() akan memberikan label pada koordinat y
- plt.title() akan memberikan judul pada histogram

```python
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori,data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Pengaturan Grid dan Colour
- Pengaturan warna pada bar bisa dilakukan pada saat pemanggilan plt.bar() dengan parameter color = '' dan alpha yang digunakan untuk pengaturan transparansi
- Pengaturan axis : y mengindikasikan bahwa garis yang ditampilkan akan berasosiasi dengan sumbu y

```python
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori,data,color='red',alpha=0.25)
plt.grid(linestyle='--',linewidth=2,axis='y',alpha=0.25)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Horisontal Bar Plot
- fungsi plt.barh() digunakan untuk membuat bar plot dalam posisi horizontal
- untuk membalikan urutan dapat dilakukan plt.barh(kategori[::-1],data[::-1])

```python
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
```

```python
plt.barh(kategori,data)
plt.grid(linestyle='--',linewidth=2,axis='x',alpha=0.25)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Stacked Bar Plot
- Parameter bottom mengindikasikan data mana yang akan dijadikan data penopang di bawah
- Legend yang muncul di hasil plottingan mengindikasikan hasil dari pemanggilan fungsi plt.legend()

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label='Data 1')
#plt.bar(kategori,data 2,label=data2)
plt.bar(kategori,data2,label='Data 2',bottom = data1)

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha = 0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh stacked bar plot')

plt.show()
```

### Grouped Bar Plot
- Berbeda dengan stacked bar plot yang menumpuk data yang di-plotting, grouped bar plot mengelompokan data yang di-plotting sesuai dengan kategorinya
- width mengindikasikan tebal/tipis dari bar plot ketika dieksekusi
- plt.xticks() berfungsi untuk mengganti label di sumbu x dengan kategori

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori = ['A','B','C','D','E']

x = np.arange(len(kategori))
x
```

```python
width=0.35

plt.bar(x-width/2,data1,width,label='Data 1')
plt.bar(x+width/2,data2,width,label='Data 2')

plt.xticks(x,kategori)

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

## Pie Plot / Pie Chart
Pembahasan ini akan mengupas cara membentuk pie chart dengan matplotlib

### Import Module

```python
#magic command untuk matplotib
%matplotlib inline
```

```python
#mengimport module yang digunakan 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
#mencetak versi matplotlib dan numpy yang sedang digunakan
print(matplotlib.__version__)
print(np.__version__)
```

### Simple Pie Plot
- parameter pertama plt.pie() akan membuat persentase dari masing-masing data dari keseluruhan data
- parame

```python
kategori = ['A','B','C','D','E']
data = [215,130,245,210,225]
```

```python
plt.pie(data,
       labels=kategori,
       autopct='%1.1f%%',
       startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```
