## Visualisasi data yang tersimpan pada pandas data frame [3]

### Import Module

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

### Sample data set

```python
df = pd.read_csv('./2020.csv')
df.head()
```

```python
df.rename(columns={
    'Country name':'Country',
    'Perceptions of corruption':'Corruption',
    'Freedom to make life choices':'Freedom'
},inplace=True)
df
```

### Bar Plot
- :5 merupakan teknik slicing dimana data yang akan diambil merupakan 5 data pertama
- Parameter kind diberi nilai bar yang mengindikasikan kita akan membentuk bar plot
- Penentuan x dan y juga diperlukan dalam parameter .plot() untuk menentukan nilai yang berasosiasi dengan sumbu x dan y

```python
df[:5].plot(x='Country',y=['Corruption','Freedom','Generosity','Social support'],kind='bar')
plt.title('Contoh Bar Plot')
plt.ylabel('Nilai')
plt.show()
```

### Line Plot

```python
df[:5].plot(x='Country',y=['Corruption','Freedom','Generosity','Social support'],kind='line')
plt.title('Contoh Line Plot')
plt.ylabel('Nilai')
plt.show()
```

### Box Plot

```python
df[:5].plot(y=['Corruption','Freedom','Generosity','Social support'],kind='box')
plt.title('Contoh Box Plot')
plt.ylabel('Nilai')
plt.xlabel('Parameter')
plt.show()
```

### Scatter Plot

```python
df.plot(x='Corruption',y='Freedom',kind='scatter')

plt.xlim((0,1))
plt.ylim((0,1))

plt.title('Contoh Scatter Plot')
plt.show()
```

### Histogram
- pengaruh nilai bins pada hasil plotting adalah semakin besar bins, maka semakin detail data yang akan dihasilkan

```python
df.plot(x='Country',y='Corruption',kind='hist',bins=10)
plt.title('Contoh Histogram')
plt.xlabel('Nilai')
plt.show()
```
