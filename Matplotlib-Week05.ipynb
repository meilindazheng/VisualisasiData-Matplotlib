```python
import matplotlib
import matplotlib.pyplot as plt
```

### Simple Scatter Plot
- Membuat data yang disimpan ke dalam x dan y
- Data point dari scatter plot akan berelasi dengan nilai di x dan y
- Menyertakan parameter label , color yang merupakan parameter opsional

```python
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
```

```python
plt.scatter(x,y,label='Data 1',color='r')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')

plt.legend()
plt.show()
```

### Multiple Scatter Plot
- Cara yang digunakan sama ketika membuat simple scatter plot
- Setiap datapoint akan berkorelasi dengan nilai di x1,y1 dan x2,y2

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2= [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x1,y1,label='Data 1',color='r')
plt.scatter(x2,y2,label='Data 2',color='b')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Multiple Scatter Plot')

plt.legend()
plt.show()
```

### Pengaturan Marker
- Pengaturan yang dapat dilakukan adalah color, linewidth, bentuk marker, edgecolor , size

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2= [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x1,y1,color='cyan',linewidth=1,marker='s',edgecolor='red',s=100)
plt.scatter(x2,y2,color='yellow',linewidth=1,marker='^',edgecolor='blue',s=200)
plt.show()
```

### Scatter  Plot untuk Data Multi Dimensi
- .keys() digunakan untuk melihat keys pada Data Frame
- .feature_name() digunakan untuk melihat feature name
- .T digunakan untuk melakukan transpose (membalikkan kolom dan baris)

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris['data'].T
```

```python
iris.keys()
```

```python
iris['data']
```

```python
iris['feature_names']
```

```python
features = iris['data'].T
features
```

```python
iris['target']
```

```python
iris['target_names']
```

```python
plt.scatter(features[0],features[1],alpha=0.2,s=100*features[3],c=iris['target'],cmap='viridis')
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.title('Iris Data Set')
plt.colorbar()
plt.show()
```
