## Belajar 3D Plot / 3D Chart

### Import Module

```python
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### 3D Axes
- Untuk membangun 3D axes, kita perlu membangun objek fig dan objek ax
- Parameter projection = 3d mengindikasikan kita akan membangun axes 3d

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.show()
```

### 3D Line Plot
- Variabel z akan menampung data 0-1 sebanyak 100 bilangan
- Parameter color mengindikasikan warna line plot pada 3D line plot
- Pengaturan label akan terlihat saat code di-run

```python
z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z * np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x,y,z,color='red')

ax.set(title='Contoh 3D line plot',xlabel = 'sumbu X',ylabel = 'sumbu Y',zlabel = 'sumbu Z')
plt.show()
```

### Scatter Plot

```python
z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z * np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x,y,z,color='red')

ax.set(title='Contoh 3D Scatter Plot',xlabel = 'sumbu X',ylabel = 'sumbu Y',zlabel = 'sumbu Z')
plt.show()
```

## Belajar Contour Plot / Contour Chart

### Import Module

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample Dataset
- Deret bilangan yang dihasilkan akan menghasilkan nilai terkecil -3 dan nilai terbesar adalah 3 serta jumlah data point adalah 100
- Fungsi np.meshgrid() digunakan untuk membentuk data area atau surface
- Variabel z akan menampung hasil akar kuadrat dari x kuadrat ditambah dengan y kuadrat

```python
n = np.linspace(-3.0,3.0,100)

x,y = np.meshgrid(n,n)
z = np.sqrt(x**2+y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

### Contour Plot
- Contour plot pada dasarnya merupakan salah satu cara bagi kita untuk memvisualisasikan data dengan format 3 dimensi kedalam format 2 dimensi
- 3 parameter yang diterima oleh contour plot akan merepresentasikan sumbu x , y dan z

```python
plt.contour(x,y,z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Filled Contour Plot

```python
plt.contourf(x,y,z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## 3D Area Plot / 3D Area Chart

### Import Module

```python
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample Dataset
- Deret bilangan yang dihasilkan akan menghasilkan nilai terkecil -3 dan nilai terbesar adalah 3 serta jumlah data point adalah 100
- Fungsi np.meshgrid() digunakan untuk membentuk data area atau surface
- Variabel z akan menampung hasil akar kuadrat dari x kuadrat ditambah dengan y kuadrat

```python
n = np.linspace(-3.0,3.0,100)

x,y = np.meshgrid(n,n)
z = np.sqrt(x**2+y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

### 3D Wireframe
- Perbedaan signifikan antara magic command Matplotlib inline dan matplotlib notebook adalah pada magic command matplotlib inline tidak memungkinkan kita untuk berinteraksi dengan wireframe hasil plotting dataset seperti pada magic command matplotlib notebook
- Perintah ax.plot_wireframe() berfungsi untuk melakukan plotting data 3D ke dalam bentuk wireframe
- Parameter ke-4 [color] merupakan parameter opsional dan berfungsi untuk mengatur warna dari hasil plottingan

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x,y,z,color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

### 3D Surface Plot 

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,color='orange')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
