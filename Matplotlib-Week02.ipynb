## [2] Membuat Plotting Sederhana

```python
#membuat plotting dengan hanya melewatkan satu parameter deret bilangan
plt.plot([2,5,7,11])
plt.ylabel('sumbu y')
#pemanggilan fungsi plt.show() digunakan untuk membuat tampilan plotting lebih rapi 
plt.show()
```

- Pemanggilan fungsi plot() digunakan dengan menyertakan sebuah deret bilangan (list/array) atau dengan menyertakan dua buah 
deret bilangan (list/array)
- Bila pemanggilan hanya menyertakan satu deret bilangan, maka nilai apda deret bilangan tersebut akan dijadikan data point pada
sumbu y, sedangkan nilai pada sumbu x akan dibuatkan sebuah deret bilangan terurut dengan nilai dimulai dari 0

```python
#mebuat plotting dengan hanya melewatkan dua parameter deret bilangan
plt.plot([2,3,4,7],[5,4,2,16])
plt.show()
```

## [3] Pengaturan Format pada Plot
- Selain menyertakan dua buah deret bilangan sebagai parameter untuk sumbu x dan y, terdapat parameter ketiga yang bisa kita lewatkan yaitu untuk mengatur warna dan jenis plotting
- Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB
- Nilai default pada parameter ketiga ini adalah 'b-'
- Pemanggilan fungsi axis () menyertakan sebuah list [xmin, xmax, ymin, ymax]

```python
#parameter pertama akan berkorelasi dengan sumbu x, parameter kedua dengan sumbu y dan pada ro (r mengindikasikan red dan o mengindikasikan bentuk)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
#pemanggilan fungsi axis() mengindikasikan  x maksimal, minimal serta y maksimal,minimal dari area plotting
plt.axis([0,6,0,20])
plt.show()
```

## [4] Multiplot Dalam Satu Pemanggilan Fungsi Plot()

```python
#menampilkan data yang dimulai dari 0. hingga maksimum 5 dengan interval antar nilai adalah 0.2
t = np.arange(0.,5.,0.2)
t
```

```python
# r merepresentasikan warna merah dan - merepresentasikan bentuknya
plt.plot(t,t,'r--')
# b merepresentasikan warna biru dan s merepresentasikan bentuk kotak (square)
plt.plot(t,t**2,'bs')
# g merepresentasikan warna hijau dan ^ merepresentasikan bentuk segitiga
plt.plot(t,t**3,'g^')
plt.show()
```

### Cara Melakukan Multiplot Dalam 1 Baris

```python
#melakukan plotting 
plt.plot(t,t,'r--',
         t,t**2,'bs',
         t,t**3,'g^')
plt.show()
```

## [5] Plotting Dengan Keywords
Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument (kwargs) seperti pada dictionary dan Pandas DataFrame

```python
#membuat dictionary sederhana dengan keys a , c, dan d
#data a dengan method arange () akan berisikan 50 angka teratur yang dimulai dari 0 dan berhenti di angka 49 (sesuai angka parameter yaitu 50)
#data c dengan method random.randint() akan berisikan 50 angka random yang dimulai dari angka 0 dan maksimal di angka 49
#data d dengan method randn() akan berisikan 50 angka yang didapat dari penghitungan standard normal distribution
data = {'a' : np.arange(50),
        'c' : np.random.randint(0,50,50),
        'd' : np.random.randn(50)}
data['b']= data['a']+10*np.random.randn(50)
data['d']= np.abs(data['d'])*100
data
```

```python
#melakukan plotting data yang bersumber dari dictionary sebelumnya dengan menggunakan plotting scatter
plt.scatter('a','b',data=data)
#memberi label entry a pada x label
plt.xlabel('entry a')
#memberi label entry b pada y label
plt.ylabel('entry b')
plt.show()
```

```python
#visualisasi point bergantung pada d dimana jika semakin besar representasi value dari d semakin besar pula size data point yang ditampilkan
#pewarnaan disesuaikan dengan c yang akan dibahas lebih lanjut
plt.scatter('a','b',c='c',s='d',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

## [6] Plotting Untuk Tipe Data Categorical
Matplotlib juga mendukung plotting untuk data dengan tipe kategori atau biasa dikenal dengan istilah categorical data categorical variabels

```python
names = ['group_a','group_b','group_c']
values = [1,10,100]
```

```python
#membuat plotting dengan menggunakan line plot
#nilai x akan berkorelasi dengan names sedangkan nilai y akan berkorelasi dengan values
plt.plot(names,values)
plt.show()
```

```python
#membuat plotting dengan jenis scatter plot
#nilai x akan berkorelasi dengan names sedangkan nilai y akan berkorelasi dengan values
plt.scatter(names,values)
plt.show()
```

```python
#membuat plotting dengan jenis bar plot
#nilai x akan berkorelasi dengan names sedangkan nilai y akan berkorelasi dengan values
plt.bar(names,values)
plt.show()
```
