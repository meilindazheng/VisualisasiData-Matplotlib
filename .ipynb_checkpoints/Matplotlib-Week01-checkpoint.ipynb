# Matplotlib

##### Matplotlib adalah salah satu module dalam pemrograman python yang dapat digunakan untuk memvisualisasikan data dalam bentuk graphs
##### Dalam Matplotlib, graphs akan digambarkan pada suatu figure, yang di dalamnya terdiri dari satu atau beberapa axes.

```python
#magic command untuk memastikan output yang dikeluarkan berada di halaman yang sama pada notebook
%matplotlib inline
```

```python
#melakukan import terhadap modul matplotlib dan numpy
import matplotlib.pyplot as plt
import numpy as np
```

## [1] Membuat plotting sederhana
membuat sebuah figure yang memiliki sebuah akses ; serta melakukan plotting data pada axes

### [1.1] Menggunakan teknik .subplots()

```python
#persiapan sample data
x = [1,2,3,4]
y = [1,4,2,3]

fig, ax = plt.subplots() #membuat sebuah figure dan axes
ax.plot(x,y)
```

### [1.2] Menggunakan teknik .plot ()

```python
#Alternatif lain
x = [1,2,3,4]
y = [1,4,2,3]
plt.plot(x,y)
```

## [2] Komponen dari sebuah Matplotlib Figure
Figure merupakan media dasar atau kontainer dasar untuk melakukan visualisasi data

Axes merupakan area gambar dan selalu berada di dalam figure | 1 figure bisa memiliki beberapa Axes

Axis mengindikasikan sumu | Axis X & Axis Y

### [2.1] Cara Membuat Figure dan Axes

#### [2.1.1] Membuat Figure Tanpa Axes

```python
# figure tanpa axes
fig = plt.figure()
```

#### [2.1.2] Membuat Figure dengan Sebuah Axes

```python
# sebuah figure dengan sebuah akses
fig, ax = plt.subplots()
```

#### [2.1.3] Membuat sebuah grid dengan sebuah axes (grid 2 x 3)

```python
# sebuah figure dengan sebuah akses dengan grid 2 x 3
# penyebutan 2 x 3 mengindikasikan 2 baris dan 3 kolom
fig, ax = plt.subplots(2,3)
```

### [2.2] Cara Menggunakan Matplotlib

#### [2.2.1] Menggunakan Teknik Object Oriented Style [OOS]
Ciri utamanya OOS :

-secara eksplisit membuat figures beserta axes, dan memanggil methods dari keduanya.

```python
# linspace digunakan untuk membuat deret angka dengan 3 parameter
# parameter pertama mengindikasikan start poin
# parameter kedua mengindikasikan ending poin
# parameter ketiga mengindikasikan jumlah data poin yang ingin dihasilkan
x = np.linspace(0,2,100)
x
```

```python
fig, ax = plt.subplots()

#plotting 3 varian data pada axes
#tanda ** mengindikasikan pemangkatan
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='Quadratic')
ax.plot(x,x**3,label='Cubic')

#menyertakan x-label dan y-label pada axes
#method .set_xlabel() dan .set_ylabel() digunakan untuk memberi label pada sumbu x dan y
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")

#menyertakan legend
ax.legend()
```

#### [2.2.2] Menggunakan Pyplot Style
Mengandalkan pyplot untuk membuat dan mengelola figures dan axes, serta menggunakan fungsi pada pyplot untuk melakukan plotting

```python
#membentuk data set 
x = np.linspace(0,2,100)
x
```

```python
#plotting 3 varian data pada axes
#tidak lagi menmbuat object figure atau axes tetapi bisa langsung melakukan plotting
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')

#menyertakan x-label dan y-label pada axes dengan .xlabel() dan .ylabel()
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")

#menyertakan legend
plt.legend()
```
