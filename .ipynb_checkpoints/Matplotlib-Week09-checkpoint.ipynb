## Matplotlib Customisation : Style Sheets

### Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Predefined Styles
- plt.style.available digunakan untuk menampilkan daftar predefined style yang terdapat pada matplotlib
- pada pembentukan contoh data set digunakan fungsi np.linspace()
- parameter yang disertakan pada linspace ada 3 
- parameter pertama mengindikasikan data terkecil, paramater kedua menunjukan data terbesar dan parameter ketiga menunjukan jumlah data point

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1,10,15)
plt.plot(x,x,'bs-',label='lin')
plt.plot(x,np.log(x),'ro-',label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Membuat Style Sendiri
- Kita dapat membuat style sendiri dengan cara membuat pengaturan di file .txt lalu menyimpannya dengan ekstensi .mplstyle
- Lalu dapat dilanjutkan dengan penggunaan style dengan cara plt.style.use()

```python
plt.style.use('./style_ku.mplstyle')
```

```python
x = np.linspace(1,10,15)
plt.plot(x,x,'bs-',label='lin')
plt.plot(x,np.log(x),'ro-',label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Menerapakan Multiple Style

```python
plt.style.use(['seaborn','dark_background'])
```

```python
x = np.linspace(1,10,15)
plt.plot(x,x,'bs-',label='lin')
plt.plot(x,np.log(x),'ro-',label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### Menerapkan temporary style

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0,2*np.pi)),'r-o')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0,2*np.pi)),'r-o')
plt.show()
```
