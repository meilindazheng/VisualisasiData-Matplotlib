### Simple 2x2 layout dengan subplots
- Langkah awal adalah dengan membentuk 2 data set yang menampung data siswa dan data nilai ujian siswa
- Lalu diikuti dengan pemanggilan fungsi subplots yang ditambah dengan parameter ncols dan nrows untuk mendefinisikan jumlah baris dan jumlah kolom 
- Parameter constrained_layout yang diberi nilai true berfungsi untuk mengatur agar tampilan antar axes memiliki jarak sehingga tampilannya tidak terlalu rapat
- Cara mengakses axes bisa dilakukan dengan pemanggilan variabel axes[x][y]  dengan x adalah index baris dan y adalah index kolom
- Sebagai catatan setiap index dimulai dari angka 0

```python
siswa = ['Cecep','Joko','Wati','Ani']
uas = [68,83,96,72]

fig,axes = plt.subplots(ncols=2,nrows=2,constrained_layout=True)

axes[0][0].bar(siswa,uas)
axes[0][1].pie(uas,labels=siswa)
axes[1][0].pie(uas,labels=siswa,explode=[0,0,0,0.2])
axes[1][1].barh(siswa,uas)

plt.show()
```

### Simple 2x2 layout dengan GridSpace [1]
- Langkah pertama adalah menghasilkan objek figure dengan memanggil fungsi plt.figure
- Selanjutnya adalah membuat objek gridspec dengan pemanggilan funsgi gridspec.GridSpec()
- Pembuatan objek axes dilakukan dengan pemanggilan fungsi objek fig.add_subplot yang akan menghasilkan axes dengan parameter gridspec yang mengindikasikan posisi dari axes

```python
fig = plt.figure(constrained_layout = True)

gs = fig.add_gridspec(ncols=2,nrows=2,figure=fig)

ax1=fig.add_subplot(gs[0,0])
ax2=fig.add_subplot(gs[0,1])
ax3=fig.add_subplot(gs[1,0])
ax4=fig.add_subplot(gs[1,1])

ax1.bar(siswa,uas)
ax2.pie(uas,labels=siswa)
ax3.pie(uas,labels=siswa,explode=[0,0,0,0.2])
ax4.barh(siswa,uas)

plt.show()
```
