
## Box Plot / Box Chart
Dalam sesi kali ini akan dipelajari cara membuat boxplot dengan matplotlib

### Import Module
- Mengimport module yang akan digunakan
- Mencetak versi dari setiap module

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample DataSet
- Membentuk random data set yang akan digunakan dalam pembelajaran

```python
np.random.seed(2)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

### Simple Box Plot
- Memanggil fungsi .boxplot akan memerlukan setidaknya 1 parameter berupa data dan parameter tambahan yang bersifat opsional
- Garis tengah akan merepresentasikan nilai median dari data yang kita miliki
- Kedua garis atas dan bawah menunjukan nilai Q1 dan Q3
- Garis diluar merepresntasikan nilai maksimal dan minimal dari data set yang divisualisasikan

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

### Notched Box Plot

```python
plt.boxplot(data,labels=['Data'],notch = True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Pengaturan Marker Untuk Outlier

```python
plt.boxplot(data,labels=['Data'],showfliers=True,flierprops={'markerfacecolor':'r','marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Horisontal Box Plot
- Cara membuat horisontal box plot adalah dengan membuat parameter vert bernilai false

```python
plt.boxplot(data,labels=['Data'],vert=False)
plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Multiple Box Plot
- Membuat 4 boxplot dalam 1 frame dengan memasukan 4 data tersebut kedalam list bernama data
- 4 data tersebut sama sama memiliki 200 data point

```python
np.random.seed(18)
data1 =  np.random.normal(100,10,200)
data2 =  np.random.normal(80,30,200)
data3 =  np.random.normal(90,20,200)
data4 =  np.random.normal(70,25,200)

data = [data1,data2,data3,data4]
labels = ['Data 1','Data 2','Data 3','Data 4']
```

```python
plt.boxplot(data,labels=labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Violin Plot atau Violin Chart
Penjelasan di bawah akan menjelaskan tentang bagaimana membentuk violin chart dengan matplotlib

### Import Module
- Mengimport module yang akan digunakan
- Mencetak versi dari setiap module

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample Data Set
- Terdapat tiga paramater untuk membentuk data set distribusi normal
- Parameter pertama adalah loc yang mengindikasikan nilai mean dari distribusi normal dataset yang dibentuk
- Parameter kedua adalah scale yang merepresentasikan standar deviasi
- Parameter ketiga adalah size yang menentukan banyak data point untuk sample data set

```python
np.random.seed(2)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

### Simple Violin Plot
- Garis yang berada di luar pada violin plot mengindikasikan nilai terbesar dan terkecil dari dataplot

```python
plt.violinplot(data)
plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Pengaturan Data pada Violin Plot
- Parameter showextrema [default=true] berfungsi untuk menampilkan garis tengah , nilai minimum hingga nilai maximum.
- Parameter showmeans merepresentasikan nilai mean dari dataset
- Parameter showmedians mengindikasikan nilai medians dari dataset
- Parameter quantiles [default=None] merepresentasikan nilai kuartil dalam dataset [Q1,Q2,Q3]

```python
plt.violinplot(data,
              showextrema=False,
              showmeans = False,
              showmedians = False,
              quantiles = None)
plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Horisontal Violin Plot
- Horisontal Violin Plot dapat diatur dengan mengubah parameter vert menjadi false

```python
plt.violinplot(data , vert = False)
plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Multiple Violin Plot

```python
np.random.seed(18)
data1 =  np.random.normal(100,10,200)
data2 =  np.random.normal(80,30,200)
data3 =  np.random.normal(90,20,200)
data4 =  np.random.normal(70,25,200)

data = [data1,data2,data3,data4]
labels = ['Data 1','Data 2','Data 3','Data 4']
```

```python
plt.violinplot(data )
plt.title('Multiple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Twin Axes
Dibawah akan dipaparkan materi untuk mempelajari Twin Axes Matplotlib

### Import Module
- Mengimport module yang akan digunakan
- Mencetak versi dari setiap modul

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Kasus Twin Axes 1
- Pada fungsi .linspace() parameter pertama akan mengindikasikan nilai terkecil
- Pada fungsi .linspace() parameter kedua akan mengindikasikan nilai terbesar
- Pada fungsi .linspace() parameter ketiga akan digunakan untuk menentukan jumlah data point yang dihasilkan oleh data frame
- Plotting pertama merupakan plotting exponensial berwarna biru dan berbentuk persegi
- Plotting kedua merupakan plotting logaritma berwarna merah dan berbentuk lingkaran

```python
x = np.linspace(1,10,25)
x
```

```python
fig,ax1 = plt.subplots()
ax1.plot(x,np.exp(x),'bs-',label='exp')
ax1.setx_label = ('Sumbu X')
ax1.sety_label = ('esp')

ax2 = ax1.twinx()
ax2.plot(x,np.log(x),'ro-',label='log')
ax1.sety_label = ('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc = 'upper left')

plt.show()
```
