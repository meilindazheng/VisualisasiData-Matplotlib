```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample DataSet
- Membentuk random data set yang akan digunakan dalam pembelajaran

```python
np.random.seed(2)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

### Simple Box Plot
- Memanggil fungsi .boxplot akan memerlukan setidaknya 1 parameter berupa data dan parameter tambahan yang bersifat opsional
- Garis tengah akan merepresentasikan nilai median dari data yang kita miliki
- Kedua garis atas dan bawah menunjukan nilai Q1 dan Q3
- Garis diluar merepresntasikan nilai maksimal dan minimal dari data set yang divisualisasikan

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

### Notched Box Plot

```python
plt.boxplot(data,labels=['Data'],notch = True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
